1.Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a url

```python
from matplotlib import pyplot as plt
import seaborn as sns

def draw_scatter_plot():
    """
       Description:
          Function is used to draw a scatter plot for a given dataset.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    sns.set_style('whitegrid')
    # Import Data
    tips_dataset = sns.load_dataset("tips")
    plt.scatter(x = "day", y = "total_bill", color = "magenta" ,data = tips_dataset)
    sns.relplot(x="total_bill", y="tip", hue="day", data = tips_dataset)
    plt.show()

if __name__ == "__main__":
    draw_scatter_plot()
```

1.Write a program to draw a violin plot of sex against​ total_bill​ for a given dataset.


```python
from matplotlib import pyplot as plt
import seaborn as sns

def draw_scatter_plot():
    """
       Description:
          Function is used to draw a violin plot for a given dataset.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    sns.set_style('whitegrid')
    # Import Data
    tips_dataset = sns.load_dataset("tips")
    sns.violinplot(x = "sex", y = "total_bill", color = "deepskyblue" ,data = tips_dataset)
    plt.show()

if __name__ == "__main__":
    draw_scatter_plot()
```
