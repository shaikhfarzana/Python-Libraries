1.Write a program to draw a scatter plot for random 1000 x and y coordinates.

```python
import plotly.express as px
import numpy as np 

def draw_scatter_plot():    
    """
       Description:
          Function is used to draw scatter plot of random number using plotly library.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

    # creating random data through randomint 
    # function of numpy.random 

    n = 1000
    random_x = np.random.randn(n)
    random_y = np.random.randn(n)

        
    plot = px.scatter(random_x, random_y, color_discrete_sequence=px.colors.qualitative.Antique)
    plot.show()

if __name__ == "__main__":
    draw_scatter_plot()
```

```python
import plotly.graph_objects as go

# Create random data with numpy
import numpy as np
np.random.seed(1)

N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5

fig = go.Figure()

# Add traces
fig.add_trace(go.Scatter(x=random_x, y=random_y0,
                    mode='markers',
                    name='markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y1,
                    mode='lines+markers',
                    name='lines+markers'))
fig.add_trace(go.Scatter(x=random_x, y=random_y2,
                    mode='lines',
                    name='lines'))

fig.show()
```
