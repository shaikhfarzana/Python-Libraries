1.Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt
from pylab import randn

def draw_scatter_plot():
   """
       Description:
          Function is used to  draw a scatter graph from random distribution in X-Y axis.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
X = randn(200)
Y = randn(200)
plt.scatter(X,Y, color='c')
plt.xticks( color = 'cyan')
plt.yticks( color = 'cyan')
plt.xlabel("X-axis", color='cyan')
plt.ylabel("Y-axis", color='cyan')
plt.title("Random Scatter Graph", color='cyan')
plt.show()


if __name__ == "__main__":
    draw_scatter_plot()
```

2.Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt 
import numpy as np 

def draw_scatter_plot_graph():
    """
       Description:
          Function is used to draw scatter plot with empty circles.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    x = np.random.randn(50) 
    y = np.random.randn(50)
    plt.scatter(x, y, s=70, facecolors='none', edgecolors='magenta')
    plt.xticks( color = 'cyan')
    plt.yticks( color = 'cyan')
    plt.xlabel("X-axis", color='cyan')
    plt.ylabel("Y-axis", color='cyan')
    plt.title("Random Scatter Graph With Empty Circles", color='cyan')
    plt.show()

if __name__ == "__main__":
    draw_scatter_plot_graph()
```
