1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

def to_plot_graph():
    """
       Description:
          Function is used to draw a line with suitable label in the x axis, y axis and a title.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    # x axis values
    x_ponits = [1,2,3]
    # y axis values
    y_points = [2,4,1]
    # Plot lines and/or markers to the Axes.
    plt.plot(x_ponits, y_points) 
    # Set the x axis label of the current axis.
    plt.xlabel('x - axis')
    # Set the y axis label of the current axis.
    plt.ylabel('y - axis')
    # Set a title 
    plt.title('This is a First Sample graph!')
    # Display a figure.
    plt.show()


if __name__ == "__main__":
    to_plot_graph()
```

2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

def to_plot_graph():
    """
       Description:
          Function is used to plot graph at given axis values with title.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    # assigning x and y coordinates
    x= [0,5,10,15,20,25]
    y = [0,1,2,3,4,5]

    # depicting the visualization
    plt.plot(x, y, color='green')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')

    # displaying the title
    plt.title("Linear graph")

    plt.show()

if __name__ == "__main__":
    to_plot_graph()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

```python
import matplotlib.pyplot as plt

def to_plot_more_than_two_lines():
    """
       Description:
          Function is used to plot two or more lines.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
   
    x1 = [10,20,30]
    y1 = [20,40,10]
    plt.plot(x1, y1, label = "line 1")
   
    x2 = [10,20,30]
    y2 = [40,10,30]
    plt.plot(x2, y2, label = "line 2")
    plt.xlabel('x - axis')
    # Set the y axis label of the current axis.
    plt.ylabel('y - axis')
    # Set a title of the current axes.
    plt.title('Two or more lines on same plot with suitable legends ')
    # show a legend on the plot
    plt.legend()
    # Display a figure.
    plt.show()


if __name__ == "__main__":
    to_plot_more_than_two_lines()
```

6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
import matplotlib.pyplot as plt

def to_customise_graph():
    """
       Description:
          Function is used to customise appearance of graph.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    x1 = [10,20,30]
    y1 = [20,40,10]

    x2 = [10,20,30]
    y2 = [40,10,30]
    # Set the x axis label of the current axis.
    plt.xlabel('x - axis')
    # Set the y axis label of the current axis.
    plt.ylabel('y - axis')
    # Set a title 
    plt.title('Two or more lines with different widths and colors with suitable legends ')
    # Display the figure.
    plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3')
    plt.plot(x2,y2, color='magenta', linewidth = 6,  label = 'line2-width-6')
    # show a legend on the plot
    plt.legend()
    plt.show()

if __name__ == "__main__":
    to_customise_graph()
```

7. Write a Python program to plot two or more lines with different styles.

```python
import matplotlib.pyplot as plt

def to_customise_line_styles():
    """
       Description:
          Function is used to plot two or more lines with different styles.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

    x1 = [10,20,30]
    y1 = [20,40,10]

    x2 = [10,20,30]
    y2 = [40,10,30]
    
    # Set the x axis label of the current axis.
    plt.xlabel('x - axis')
    # Set the y axis label of the current axis.
    plt.ylabel('y - axis')
    # Plot lines and/or markers to the Axes.
    plt.plot(x1,y1, color='cyan', linewidth = 5,  label = 'line1-dotted',linestyle='dotted')
    plt.plot(x2,y2, color='magenta', linewidth = 8,  label = 'line2-dashed', linestyle='dashdot')

    # Set a title 
    plt.title("Plot with two or more lines with different styles")
    # show a legend on the plot
    plt.legend()
    # function to show the plot
    plt.show()

if __name__ == "__main__":
    to_customise_line_styles()
```

8. Write a Python program to plot two or more lines and set the line markers.

```python
import matplotlib.pyplot as plt

def to_plot_with_markers():
    """
       Description:
          Function is used to plot two or more lines and set line markers.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    x = [1,4,5,6,7]
    y = [2,6,3,6,3]

    # plotting the points 
    plt.plot(x, y, color='cornflowerblue', linestyle='dashdot', linewidth = 5,
            marker='v', markerfacecolor='coral', markersize=15)
    #Set the y-limits of the current axes.
    plt.ylim(1,8)
    #Set the x-limits of the current axes.
    plt.xlim(1,8)

    # naming the x axis
    plt.xlabel('x - axis')
    # naming the y axis
    plt.ylabel('y - axis')

    # giving a title to my graph
    plt.title('Display marker')
    # function to show the plot
    plt.show()

if __name__ == "__main__":
    to_plot_with_markers()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
import matplotlib.pyplot as plt

def to_set_new_axis_values():
    """
       Description:
          Function is used to set new axis values.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    X = range(1, 50)
    Y = [value * 3 for value in X]
    plt.plot(X, Y)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Draw a line.')
    # shows the current axis limits values
    print(plt.axis())
    # set new axes limits
    # Limit of x axis 0 to 100
    # Limit of y axis 0 to 200
    plt.axis([0, 100, 0, 200]) 
    # Display the figure.
    plt.show()

if __name__ == "__main__":
    to_set_new_axis_values()
```
