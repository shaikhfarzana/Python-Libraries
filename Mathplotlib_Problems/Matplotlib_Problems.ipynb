1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt

def to_plot_graph():
    """
       Description:
          Function is used to draw a line with suitable label in the x axis, y axis and a title.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    # x axis values
    x_points = [1,2,3]
    # y axis values
    y_points = [2,4,1]
    # Plot lines and/or markers to the Axes.
    plt.plot(x_points, y_points) 
    plt.xticks( color = 'cyan')
    plt.yticks( color = 'cyan')
    

    # Set the x axis label of the current axis.
    plt.xlabel('x-axis', color= 'cyan')
    # Set the y axis label of the current axis.
    plt.ylabel('y-axis', color = 'cyan')
    # Set a title 
    plt.title('This is a First Sample graph!', color='cyan')
    # Display a figure.
    plt.show()


if __name__ == "__main__":
    to_plot_graph()
```

7. Write a Python program to plot two or more lines with different styles.

```python
import matplotlib.pyplot as plt

def to_customise_line_styles():
    """
       Description:
          Function is used to plot two or more lines with different styles.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

    x1 = [10,20,30]
    y1 = [20,40,10]

    x2 = [10,20,30]
    y2 = [40,10,30]
    
    # Set the x axis label of the current axis.
    plt.xlabel('x - axis', color='cyan')
    # Set the y axis label of the current axis.
    plt.ylabel('y - axis', color='cyan')
    # Plot lines and/or markers to the Axes.
    plt.plot(x1,y1, color='cyan', linewidth = 5,  label = 'line1-dotted',linestyle='dotted')
    plt.plot(x2,y2, color='magenta', linewidth = 8,  label = 'line2-dashed', linestyle='dashdot')
    plt.xticks( color ='cyan')
    plt.yticks( color ='cyan')
   

    # Set a title 
    plt.title("Plot with two or more lines with different styles", color='cyan')
    # show a legend on the plot
    plt.legend()
    # function to show the plot
    plt.show()

if __name__ == "__main__":
    to_customise_line_styles()
```

8. Write a Python program to plot two or more lines and set the line markers.

```python
import matplotlib.pyplot as plt

def to_plot_with_markers():
    """
       Description:
          Function is used to plot two or more lines and set line markers.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    x = [1,4,5,6,7]
    y = [2,6,3,6,3]

    # plotting the points 
    plt.plot(x, y, color='cornflowerblue', linestyle='dashdot', linewidth = 5,
            marker='v', markerfacecolor='coral', markersize=15)
    plt.xticks( color = 'cyan')
    plt.yticks( color = 'cyan')
    plt.xlabel('x-axis', color= 'cyan')
    plt.ylabel('y-axis', color = 'cyan')
    #Set the y-limits of the current axes.
    plt.ylim(1,8)
    #Set the x-limits of the current axes.
    plt.xlim(1,8)
    # giving a title to my graph
    plt.title('Display marker', color='cyan')
    # function to show the plot
    plt.show()

if __name__ == "__main__":
    to_plot_with_markers()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python
import matplotlib.pyplot as plt

def to_set_new_axis_values():
    """
       Description:
          Function is used to set new axis values.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """
    X = range(1, 50)
    Y = [value * 3 for value in X]
    plt.plot(X, Y)
    plt.xticks( color = 'cyan')
    plt.yticks( color = 'cyan')
    plt.xlabel('x-axis', color= 'cyan')
    plt.ylabel('y-axis', color = 'cyan')
    plt.title('Draw a line.', color='cyan')
    # set new axes limits
    # Limit of x axis 0 to 100
    # Limit of y axis 0 to 200
    plt.axis([0, 50, 0, 200]) 
    plt.grid(True)
    # Display the figure.
    plt.show()

if __name__ == "__main__":
    to_set_new_axis_values()
```

12. Write a Python program to create multiple types of charts.

```python
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num

def to_create_multiple_type_charts():
    """
       Description:
          Function is used to create to create multiple types of charts.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017)]

x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]

fig = plt.figure()

graph = fig.add_subplot(111)

# Plot the data as a red line with round markers
graph.plot(x,y,'r-p')

# Set the xtick locations
graph.set_xticks(x)

# Set the xtick labels
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data]
        )
plt.xlabel('x-axis', color= 'cyan')
plt.ylabel('y-axis', color = 'cyan')
plt.xticks( color = 'cyan')
plt.yticks( color = 'cyan')
plt.show()

if __name__ == "__main__":
        to_create_multiple_type_charts()

```

13.  Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue.

```python
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num

def to_customise_grid_lines():
   """
       Description:
          Function is used to customise grid lines.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

   data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
         (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
         (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
         (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
         (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

   x = [date2num(date) for (date, value) in data]
   y = [value for (date, value) in data]

   fig = plt.figure()

   graph = fig.add_subplot(111)

   # Plot the data as a red line with round markers
   graph.plot(x,y,'r-o')

   # Set the xtick locations
   graph.set_xticks(x)

   # Set the xtick labels
   graph.set_xticklabels(
         [date.strftime("%Y-%m-%d") for (date, value) in data]
         )

   plt.title('Closing stock value of Alphabet Inc.', color='cyan') 
   # Customize the grid
   plt.grid(linestyle='--', linewidth='2.0', color='cyan')
   plt.xlabel('Date', color= 'cyan')
   plt.ylabel('Closing value', color = 'cyan')
   plt.xticks( color = 'cyan')
   plt.yticks( color = 'cyan')
   plt.show()

if __name__ == "__main__":
    to_customise_grid_lines()
```

14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks.

```python
import datetime as DT
from matplotlib import pyplot as plt
from matplotlib.dates import date2num

def to_customise_grid_lines():
   """
       Description:
          Function is used to customise major and minor grid lines.
       Parameter:
          No parameter required.
       Return:
          Returns nothing.
    """

   data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
         (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
         (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
         (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
         (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

   x = [date2num(date) for (date, value) in data]
   y = [value for (date, value) in data]

   fig = plt.figure()

   graph = fig.add_subplot(111)

   # Plot the data as a red line with round markers
   graph.plot(x,y,'r-o')

   # Set the xtick locations
   graph.set_xticks(x)

   # Set the xtick labels
   graph.set_xticklabels(
         [date.strftime("%Y-%m-%d") for (date, value) in data]
         )

   plt.title('Closing stock value of Alphabet Inc.', color='cyan') 
   plt.xlabel('Date', color= 'cyan')
   plt.ylabel('Closing value', color = 'cyan')
   plt.xticks( color = 'cyan')
   plt.yticks( color = 'cyan')
   # Customize the grid
   # Turn on the minor TICKS, which are required for the minor GRID
   plt.minorticks_on()

    # Customize the major grid
   plt.grid(which='major', linestyle='--', linewidth='2.0', color='cyan')
    # Customize the minor grid
   plt.grid(which='minor', linestyle=':', linewidth='1.0', color='blueviolet')

    # Turn off the display of all ticks.
   plt.tick_params(which='both', # Options for both major and minor ticks
                    top='off', # turn off top ticks
                    left='off', # turn off left ticks
                    right='off',  # turn off right ticks
                    bottom='off') # turn off bottom ticks
   plt.show()

if __name__ == "__main__":
    to_customise_grid_lines()
```

3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.

```python
from  matplotlib import pyplot as plt

def to_take_axis_values_from_textfile():
    with open("test.txt") as f:
        data = f.read()
    data = data.split('\n')
    # axis values are taken from textfile.
    x = [row.split(' ')[0] for row in data]
    y = [row.split(' ')[-1] for row in data]

    plt.plot(x, y)
    plt.xlabel('x-axis', color= 'cyan')
    plt.ylabel('y-axis', color = 'cyan')
    plt.xticks( color = 'cyan')
    plt.yticks( color = 'cyan')
    plt.title('Taking Axis Values From Textfile', color='cyan')
    plt.show()

if __name__ == "__main__":
    to_take_axis_values_from_textfile()
```
